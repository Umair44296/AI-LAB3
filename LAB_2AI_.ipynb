```python
# append() - Adds an element at the end of the list
fruits = ["Apple", "Mango", "Banana"]
fruits.append("Orange")
print("After append:", fruits)
```

```python
# clear() - Removes all the elements from the list
fruits = ["Apple", "Mango", "Banana"]
fruits.clear()
print("After clear:", fruits)
```

```python
# copy() - Returns a copy of the list
fruits = ["Apple", "Mango", "Banana"]
fruits2 = fruits.copy()
print("Original list:", fruits)
print("Copied list:", fruits2)
```

```python
# count() - Returns the number of elements with the specified value
fruits = ["Apple", "Mango", "Apple", "Banana", "Apple"]
print("Count of Apple:", fruits.count("Apple"))
```

```python
# extend() - Adds elements of another list to the end of the current list
fruits = ["Apple", "Mango"]
veggies = ["Carrot", "Potato"]
fruits.extend(veggies)
print("After extend:", fruits)
```

```python
# index() - Returns the index of the first element with the specified value
fruits = ["Apple", "Mango", "Banana"]
print("Index of Mango:", fruits.index("Mango"))
```

```python
# insert() - Adds an element at the specified position
fruits = ["Apple", "Mango", "Banana"]
fruits.insert(1, "Orange")
print("After insert:", fruits)
```

```python

# pop() - Removes the element at the specified position
fruits = ["Apple", "Mango", "Banana"]
fruits.pop(1)
print("After pop:", fruits)
```

```python
# remove() - Removes the item with the specified value
fruits = ["Apple", "Mango", "Banana"]
fruits.remove("Mango")
print("After remove:", fruits)
```

```python
# reverse() - Reverses the order of the list
fruits = ["Apple", "Mango", "Banana"]
fruits.reverse()
print("After reverse:", fruits)
```

```python
# sort() - Sorts the list
fruits = ["Banana", "Apple", "Mango", "Cherry"]
fruits.sort()
print("After sort:", fruits)
```
