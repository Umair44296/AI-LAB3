```python
# DataType Output: str
x = "A"
print("Type of x:", x, type(x))
# DataType Output: str
x = "Hello World"
print("Type of x:", x, type(x))
# DataType Output: int
x = 50
print("Type of x:", x, type(x))
# DataType Output: float
x = 60.5
print("Type of x:", x, type(x))
# DataType Output: bool
x = True
print("Type of x:", x, type(x))
# DataType Output: complex
x = 3j
print("Type of x:", x, type(x))
# DataType Output: list
x = ["item1", "item2", "item3"]
print("Type of x:", x, type(x))
# DataType Output: tuple
x = ("Father", "Mother")
print("Type of x:", x, type(x))
# DataType Output: range
x = range(5)
print("Type of x:", x, type(x))
# DataType Output: dict
x = {"name": "Ali", "age": 24}
print("Type of x:", x, type(x))
# DataType Output: set
x = {"Apple", "Mango", "Banana"}
print("Type of x:", x, type(x))
```

```python
#Get two input values from user and calculate sum
num1 = int(input('Your integer 1 : '))
num2 = int(input('Your integer 2 : '))
sum = num1+num2
print('Sum is :', sum)
```

```python
# Arithmetic Operators
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Exponent:", a ** b)
```

```python
# Assignment Operators
x = 10
print("x =", x)

x += 5
print("x += 5 :", x)

x -= 3
print("x -= 3 :", x)

x *= 2
print("x *= 2 :", x)

x /= 4
print("x /= 4 :", x)

x //= 2
print("x //= 2 :", x)

x %= 3
print("x %= 3 :", x)

x **= 2
print("x **= 2 :", x)
```

```python
# Comparison Operators
a = 10
b = 5

print("a == b :", a == b)
print("a != b :", a != b)
print("a > b  :", a > b)
print("a < b  :", a < b)
print("a >= b :", a >= b)
print("a <= b :", a <= b)
```

```python
# Logical Operators
a = True
b = False

print("a and b :", a and b)
print("a or b  :", a or b)
print("not a   :", not a)
print("not b   :", not b)
```
